```python
import pandas as pd
import json

with open('../DATA/memory_cpu/prometheus/cpu2.json', 'r') as f:
  data = json.load(f)
```

```python
cont = 0
df = pd.DataFrame()

for res in data["result"]:
    pod_name = res["metric"]["pod"]
    val = res["values"]
    temp = pd.DataFrame(val, columns=["Time", pod_name])
    temp.set_index("Time", drop=True, inplace=True)
    df = pd.concat([df, temp], axis="columns")

df.head(500)
```

```python
df.to_csv("../DATA/memory_cpu/processed/cpu_usage2.csv")
```
